# DATA PREPROCESSING

```python
# Deleting columns with missing values greater than 80
data = data.drop((missing[missing['Total'] > 81]).index,axis=1) 
data = data.drop(data.loc[data['Electrical'].isnull()].index) 
```

```python
# Droping data with weak correlation
data = data.drop(correlations.iloc[21: , 0].values,axis=1)
```

```python
# Dealing with missing values
cols_name = data.drop(columns="Id").select_dtypes(include=['number']).columns
for col in cols_name :
    NaN_ = data[col].isnull() 
    samples = data[col].dropna().sample(n=NaN_.sum(), replace=True) 
    data.loc[NaN_, col] = samples.values
```

```python
data.drop(columns="Id").select_dtypes(include=['number']).isnull().sum().sort_values(ascending=False)
```

```python
cols_name = data.drop(columns="Id").select_dtypes(include=['object']).columns

for col in cols_name :
      mode_ = data[col].mode()[0]
      data[col].fillna(mode_, inplace=True)

data.drop(columns="Id").select_dtypes(include=['object']).isnull().sum().sort_values(ascending=False)
```

```python
# Dealing with outliers
def out_liers(data , col):
    if pd.api.types.is_numeric_dtype(data[col]):
        Q1  , Q3 = np.percentile(data[col] , [25,75])
        IQR = Q3 - Q1
        l_bound = Q1 - (1.5 * IQR) ## Lower Bound
        u_bound = Q3 + (1.5 * IQR) ## Upper Bound
        data = data[(data[col]>=l_bound) & (data[col] <=u_bound)]

    return data
```

```python
for col in data.columns:
    data = out_liers(data, col)
```

```python
##creating train/test sets
from sklearn.model_selection import  train_test_split

X = data.iloc[:, :-1].values
y = data.iloc[:, -1:].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
#scaling using standard scaler
sc = StandardScaler(with_std=True)
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# MODEL BUILDING AND EVALUATION

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import learning_curve
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# creating an evaluation function
def evaluate(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f"Mean Squared Error: {mse}")
    print(f"Root Mean Squared Error: {rmse}")
    print(f"Mean Absolute Error: {mae}")
    print(f"R-squared: {r2}")
    
    # Learning curve
    train_sizes, train_scores, test_scores = learning_curve(model, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
    train_scores_mean = -np.mean(train_scores, axis=1)
    test_scores_mean = -np.mean(test_scores, axis=1)
    plt.figure(figsize=(8, 6))
    plt.plot(train_sizes, train_scores_mean, label='Train score')
    plt.plot(train_sizes, test_scores_mean, label='Validation score')
    plt.xlabel('Training examples')
    plt.ylabel('Mean Squared Error')
    plt.title('Learning Curve')
    plt.legend()
    plt.show()
```

### LINEAR REGRESSION

```python
lr_reg = linear_reg = LinearRegression()
```

```python
evaluate(lr_reg, X_train, X_test, y_train.ravel(), y_test.ravel())
```

### RANDOM FOREST

```python
rf_reg = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
evaluate(rf_reg, X_train, X_test, y_train.ravel(), y_test.ravel())
```

### XGBoost

```python
xgb_regressor = XGBRegressor(n_estimators=100, random_state=42)
```

```python
evaluate(xgb_regressor, X_train, X_test, y_train.ravel(), y_test.ravel())
```

#### >Random Forest Regressor and Linear Regression performed relatively similarly in terms

#### performed reasonably well.
#### >R-squared values for all models are around 0.87, indicating that approximately 87% of

# HYPERPARAMETER TUNING RANDOM FOREST

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_reg = RandomForestRegressor(random_state=42)
```

```python
# selecting parameters for hyperparameter tuning
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# performing grid search using cross-validation
grid_search = GridSearchCV(estimator=rf_reg, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train.ravel())
```

```python
# Best parameters and best score
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", -grid_search.best_score_)
```
